```python
import pandas as pd
```

```python
stats = pd.read_csv("TEDViz.csv", index_col="Date", parse_dates=True)
```

```python
stats.head()
```

```python
stats["TED_Views"] = stats["TED_Views"].str.replace(',', '')
```

```python
stats.head()
```

```python
import plotly
from plotly.graph_objs import Scatter, Layout
import plotly.graph_objs as go


literate = [go.Bar(x=stats.index, y=stats["TED_Views"])]


plotly.offline.plot({ 'data': literate,
            'layout': {
               'title': 'POPULATION LITERACY RATE IN TN',
               'xaxis': {
                 'title': 'YEAR'},
               'yaxis': {
                'title': 'POPULATION (in lakhs) '}
        }})
```
